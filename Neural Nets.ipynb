```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense,BatchNormalization
```

```python
mnist = tf.keras.datasets.mnist

(x_train,y_train),(x_test,y_test) = mnist.load_data()
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
def UniformNet():
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Flatten())
    model.add(Dense(50,activation="relu", 
                    kernel_initializer='glorot_normal'))
    model.add(BatchNormalization()) # Batch Normalization (increased accuracy by 7% on train set)
    # model.add(tf.keras.layers.Dropout(0.3)) # Dropout (Decreased accuracy by 6% on train set)
    model.add(Dense(109, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model.add(Dense(70, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model.add(Dense(109, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model.add(Dense(70, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model.add(tf.keras.layers.Dropout(0.3)) # Dropout (Decreased accuracy by 6% on train set)
    model.add(Dense(109, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model.add(Dense(70, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model.add(Dense(109, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model.add(Dense(70, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model.add(Dense(12, activation="softmax",
                    kernel_initializer='glorot_normal'))
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    model.fit(x_train,y_train,epochs = 10)
    model.summary()
    
UniformNet()
```

```python
def PyramidNet():
    model1 = tf.keras.models.Sequential()
    model1.add(tf.keras.layers.Flatten())
    model1.add(Dense(50,activation="relu", 
                    kernel_initializer='glorot_normal'))
    model1.add(tf.keras.layers.Dropout(0.3)) # Dropout
    model1.add(Dense(600, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model1.add(BatchNormalization()) # Batch Normalization
    model1.add(Dense(25, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model1.add(Dense(289, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model1.add(Dense(13, activation = "relu",
                    kernel_initializer='glorot_normal'))
    # model1.add(tf.keras.layers.Dropout(0.3)) # Dropout
    model1.add(Dense(134, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model1.add(Dense(7, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model1.add(Dense(67, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model1.add(Dense(4, activation = "relu",
                    kernel_initializer='glorot_normal'))
    model1.add(Dense(34, activation="softmax",
                    kernel_initializer='glorot_normal'))
    model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    model1.fit(x_train,y_train,epochs = 10)
    model1.summary()

PyramidNet()
```
